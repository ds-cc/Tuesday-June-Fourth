```python
# load

import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/ds-cc/Tuesday-June-Fourth/master/survey.csv')

df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df = df.rename(columns={'Unnamed: 0': 'Name'})
```

```python
df = df.rename(columns={'What is your programming level?': 'Programming',
                        'What is your math level?': 'Math',
                       'What level are you going to be when you finish Data Science at Lambda School? ðŸ˜€': 'Future level'
                       })

df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('dark_background')
df[['Programming', 'Math']].plot()
```

```python
df['In a classroom I'] = df['In a classroom I'].apply(lambda x: 1 if x == 'ask a lot of questions' else 0)
```

```python
# 0 is not asking a lot of questions

df['In a classroom I'].plot.hist()
```

```python
df[df['In a classroom I']==1].describe()
```

```python
df[df['In a classroom I']==0].describe()
```

We found that the majority of students reported higher math ability than programming. Also, aside from one outlier, all self reported their future ability to be either 9 or 10. 

There is no correlation between what level people said they would be in the future and classroom participation.

The mean of self reported math, programming, and future abilities was higher for those students who reported preferring sitting and taking notes.
